# Preprocessing

```python
current_dir = os.listdir()
```

```python
temp_ratings = pd.read_csv('ratings.txt', header=None)
```

```python
current_dir[0]
```

```python
eeg_data = []

for ftemp in current_dir:
    if ftemp[-3:] == 'csv':
        tgt_partic = int(ftemp[0:2])
        #print(tgt_partic)
        temprat = temp_ratings[temp_ratings[0] == tgt_partic]
       # print(temprat)
        if len(temprat)==0:
            tgt_partic_rat = np.array([[0,0,0]])
        else:
            tgt_partic_rat = temprat.values
            print(tgt_partic_rat)
            
        
        dftemp = pd.read_csv(ftemp, header='infer')
        dftemp['subj'] = tgt_partic
        
        dftemp['hilo'] = ftemp[3:5]
        if ftemp[3:5] == 'hi':
            dftemp['subjective_rating'] = tgt_partic_rat[0][2]
            #print(tgt_partic_rat[0][2])
            #print(dftemp)
        else:
            dftemp['subjective_rating'] = tgt_partic_rat[0][1]
        eeg_data.append(dftemp)
        
eeg_data = pd.concat(eeg_data)
#eeg_data.info()
```

# SVM CLASSIFIERS

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import cross_val_score
from sklearn.metrics import plot_confusion_matrix
```

```python
#LStandardScalerar, Polynomial, RBF, Sigmoid
#kernel coefficient gamma
#2 units of penalty
svm_classifier = svm.SVC(kernel= 'rbf', gamma='auto', C = 2 )
svm_classifier.fit(X_train, y_train)

y_predict = svm_classifier.predict(X_test)
```

```python
cross_val_score(svm_classifier, X_train, y_train, cv=5).mean()
```

```python
class_names = list(set(y))
```

```python
disp = plot_confusion_matrix(svm_classifier, X_test, y_test,
                                 display_labels=class_names,
                                 cmap=plt.cm.Blues)
```

# KNN

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)

y_predict = knn_classifier.predict(X_test)

cross_val_score(knn_classifier, X_train, y_train, cv=5).mean()
```

```python
print(confusion_matrix(y_test, y_predict))
print(classification_report(y_test, y_predict))

class_names = list(set(y))

disp = plot_confusion_matrix(knn_classifier, X_test, y_test,
                                 display_labels=class_names,
                                 cmap=plt.cm.Blues)
```
